```
#creating the list containing the string only
import pandas as pd
record_list=[] #for storing the features 
y=[] #for storing the target

filepath = '/content/drive/MyDrive/Colab Notebooks/dataFiles/DNA_Genome_Analysis_covid19/sequences.fasta'
records = list(SeqIO.parse(filepath , 'fasta'))

for i in range(len(records)):
  record_list.append(str(records[i].seq))
  y.append(1)

print("The lenth of the dataset list---->",len(record_list))
```

```
#creating it as a pandas DataFrame
data = pd.DataFrame({'DNASequence' : record_list , 'status' : y})
data.head(4) 
```
